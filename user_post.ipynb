```python
import latlon as LatLon
from latlon import *
from langdetect import *

user='aymantounsi'
users=[]
users.append(user)

#Prétraitement
profile = instaloader.Profile.from_username(loader.context,user)
context=instaloader.InstaloaderContext(sleep=True, quiet=False, user_agent=None, max_connection_attempts=3, request_timeout=300.0, rate_controller=None, fatal_status_codes=None, iphone_support=True)

#Informations de profile
_username=profile.username
_mediacount=(profile.mediacount)
_followers=(profile.followers)
_followees=(profile.followees)
#Informations de poste
posts =profile.get_posts()
_postsliste=[]
_postsliste=list(posts)
nombre_de_postes=len(_postsliste)
shortcode=[]
for i in _postsliste :
    a=(str(i)[6:-1])
    shortcode.append(a)
print(shortcode)
liste_1=[]
for i in (shortcode) :
    Liste=[]
    Liste.append(_username)
    Liste.append(_followers)
    Liste.append(_followees)
    Liste.append(_mediacount)

    post=Post.from_shortcode(loader.context,i)
    #Likes
    likes=post.likes
    Liste.append(likes)
    #Comments
    comments=(post.comments)
    Liste.append(comments)
    #Location
    distance =''
    place=''
    location=post.location
    print(location)
    corrd=[]
    if ((location)==None):
        distance='Nul'        
    else:
        for i in (location) :
            a=(str(i))
            corrd.append(a)
            print(corrd)
        lat=corrd[-2]
        lng=corrd[-1]
        location_name=corrd[1]
        if ((lat!='')&(lng!='')):
            loc_found=LatLon(Latitude(lat), Longitude(lng))
            ile_de_France = LatLon(Latitude(48.8499), Longitude( 2.6370))   
            if ((ile_de_France.distance(loc_found))<=50):
                distance='In Zone'
                    
            else:
                distance='Off Zone'
        if(location_name!=''):
            place=location_name
        else:
            distance='Nul'
            place='Nul'
    Liste.append(distance)    
    Liste.append(place)
    #Caption
    caption=post.caption
    Liste.append(caption)
    try:
        print(caption)
        #Caption langue
        langue_caption=detect((caption))
    except :
        langue_caption = 'undefined'
    Liste.append(langue_caption)
    print(langue_caption)

    #Title
    title=post.title
    Liste.append(title)
    #URL
    url=post.url
    Liste.append(url)
    #Date
    date=post.date_local
    Liste.append(date)
    #Liste des Likes
    L= post.get_likes()
    Liste.append(L)
    #Liste des Commentaires
    C=post.get_comments()
    Liste.append(C)
    #Tagged users
    tagged_users=post.tagged_users
    Liste.append(tagged_users)
    #Sponsorisée
    sponsored=post.is_sponsored
    Liste.append(sponsored)
    #The Post’s sponsors
    sponsor_users=post.sponsor_users
    Liste.append(sponsor_users)
    #Number of sponsors
    taille=len(sponsor_users)
    Liste.append(taille)
    #Video viewers
    video_view_count=post.video_view_count
    Liste.append(video_view_count)
    #Engagement Rate
    engagement_rate=(post.likes/_followers)*100
    Liste.append(engagement_rate)
    #Ajout dans la liste principale
    liste_1.append(Liste)
df=pd.DataFrame(liste_1,columns=['user','Followers','Followees','mediacount','Likes','Comments','Distance','place','Caption','langue','Title','URL','Date','Liste des Likes','Liste des Commentaires','Tagged users','Sponsorisée','The Posts sponsors','Number of sponsors','Video viewers','Engagement Rate'])
#Création d'un fichier csv pour stocker les données
path=['C:/Users/User/Desktop/orthlane/user_post/']
ext=['.csv']
a=path[0]+users[0]+ext[0]
df.to_csv(a, index=False)
df
```
