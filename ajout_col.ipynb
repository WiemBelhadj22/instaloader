```python
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\User\Desktop\orthlane\data_final\bisous.csv')
```

```python
#Effacer les redendances
df.drop_duplicates(subset="Username",inplace=True)
```

```python
#Ajouter une colone de langue et appender les r√©sultats
df['langue']=''
from langdetect import detect
def langue(x):
    try:
        lang = detect(x)
    except:
        lang = 'Other'
    return lang
for i in df.index:
    a=langue(df['bio'][i])
    df['langue'][i]=a
    print(df['langue'][i]) 
```

```python
df['Engagement Rate']=0.
```

```python
#Visualiser les Followers/Followees
fig,ax=plt.subplots()
df_2=pd.DataFrame({'Followers':list(df['Followers']),'Followees':list(df['Followees'])},index=list(df['Username']))
df_2.plot.bar(figsize =(30,30),ax=ax)
```

```python
#Visualiser les Followers/Followees/Number of posts
fig,ax=plt.subplots()
df_3=pd.DataFrame({'Number of posts':list(df['Number of posts']),'Followers':list(df['Followers']),'Followees':list(df['Followees'])},index=list(df['Username']))
df_3.plot.bar(figsize =(30,30),ax=ax)
```
